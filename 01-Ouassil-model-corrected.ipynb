```python
# Basic libraries
import pandas as pd
import random

# Model import 
from class_rfc import ORFC, ORFC_no_smote

```

```python
random.seed(10)
random_seeds = random.sample(range(1, 1000000), 10)
```

```python
print(random_seeds)
```

```python
# Step 1: Import the data
# Base data for contructing and training the models
# plm = Blood plasma data
# csf = Cerebrospinal fluid data
plm_data = pd.read_excel("data/"+'a1-gt15_plasma_features_names_biopy_gravy.xlsx', header=0, index_col=0)
csf_data = pd.read_excel("data/"+'a2-gt15_csf_features_names_biopy_gravy.xlsx', header=0,index_col=0)

# Merge plasma and csf data
total_data = pd.concat([plm_data, csf_data], axis=0, ignore_index=True).fillna(0).drop(['mass'], axis=1)
original_model_test = []
```

```python
# Step 2: Create the model and train it
# Execute 10 times with different random seeds
for seed in random_seeds:
    Original_model = ORFC(total_data, k=38, random_state=seed)
    Original_model.rmetrics()
    X_new_data = pd.read_excel("data/z0-data_100_prot.xlsx", header=0, index_col=0)
    original_model_test.append(Original_model.mkpred(X_new_data))
```
